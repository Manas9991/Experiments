Counting occurences:

```
words = txt.split(' ') #Converts the text into a list of words
wordsu = [] #store unique words for counting

for i in words:
  if (i not in wordsu):
    wordsu.append(i)

count = [] #store count of all unique words

for i in wordsu: #for each word
  occ = 0 #initialize occurences
  for j in words: #go through the text looking for that word
    if (i==j): #if word if found, increment occurence
      occ += 1
  count.append(occ) #store the number of occurences for that word

result = dict(zip(wordsu, count)) #link the words and their counts
result = {k: v for k, v in sorted(result.items(), key=lambda item: item[1], reverse = True)} #sorting according to occurences
```

Plotting occurences:

```
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(20,10))
plt.title('Number of occurences of each unique word in the text', fontsize = 16)
plt.ylabel('Number of occurences', fontsize = 12)
plt.xlabel('Words', fontsize = 12)
plt.xticks(rotation = "vertical")

keys = result.keys()
values = result.values()

plt.bar(keys, values)
plt.show()
```

```
sum20 = sum(count[0:20]) #number of 20 most commonly used words
sumtotal = len(words) #total number of words

percent = 100.0*(sumtotal-sum20)/sumtotal

print("The number of 20 most commonly used words is: ", sum20, "\nThe total number of words is: ", sumtotal)
print("The percentage of 20 most commonly used words to total is: ", round(percent,2))
```
